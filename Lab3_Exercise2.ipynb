```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sklearn as sk
```

```python
bostonHouseFrame = pd.read_csv('s3://rmit-ml-2020s1-lab-data/BostonHousingPrice/housing.data.csv', delimiter = '\s+')
bostonHouseFrame
```

```python
X = bostonHouseFrame[['RM']]
Y = bostonHouseFrame[['MEDV']]
```

```python
from sklearn.model_selection import train_test_split

trainX, testX, trainY, testY = train_test_split(np.array(X), np.array(Y), test_size = 0.2)
print(trainX.shape)
print(testX.shape)
print(trainY.shape)
print(testY.shape)
```

```python
from sklearn import linear_model
linReg = linear_model.LinearRegression()
```

```python
linReg.fit(trainX, trainY)
```

```python
print(linReg.intercept_)
print(linReg.coef_)
```

```python
predYRm = linReg.predict(testX)
```

```python
from sklearn.metrics import mean_squared_error
print('Mean squared error ', mean_squared_error(testY, predYRm))
```

```python
plt.rcParams['figure.figsize'] = [20, 10]
plt.figure()

plt.subplot(1, 2, 1)
plt.scatter(trainX, trainY)
plt.plot(trainX, linReg.predict(trainX), color = 'red', linewidth = 7.0)
plt.xlabel("RM")
plt.ylabel("MDEV")
plt.title('Train Data')

plt.subplot(1, 2, 2)
plt.scatter(testX, testY)
plt.plot(testX, linReg.predict(testX), color = 'blue', linewidth = 7.0)
plt.xlabel("RM")
plt.ylabel("MDEV")
plt.title("Test Data")

plt.show()
```
